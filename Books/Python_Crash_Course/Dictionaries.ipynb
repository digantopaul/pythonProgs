```python
#Looping through all the keys in a dictionary, keys()
fav_langs = {
    "jen" : "english",
    "deepa" : "hindi",
    "sussane" : "german",
    "brad" : "deutsch",
}

print("The participants are ")
for name in fav_langs.keys():
    #print(name.title(),sep=",",end=" ")     #Here sep="," will not work as sep needs multiple objects to put in the sep string. With
    #single object being printed every time, sep doesnt come into play.
    print(f"{name.title()}, ", end = "")
print("\n\n")

#Looping through the dictionary by defaults loops through the keys of the dictionary. Using th keys() method attributes to writing
#clean and understandable code writing. 
# Example 1
for name in fav_langs:      #Here you can see the keys() method is not called, yet the variable name iterated over the keys of fav_langs.
    print(f"hi {name.title()}!")

# Example 2
frnds = ["deepa", "brad", "erin"]
for name in fav_langs:
    print(f"hi {name.title()}!")
    if name in frnds:
        print(f"{name.title()}, I see you have taken {fav_langs[name].title()}.")
    if name not in fav_langs.keys():
        print(f"{name.title()}, please take the poll!")

```

```python
fav_langs = {
    "jen" : "english",
    "deepa" : "hindi",
    "sussane" : "german",
    "brad" : "deutsch",
    "demon" : "english",
    "jeremy" : "deutsch",
    "rishabh" : "hindi",
    "saber" : "deutsch",
}
#To get unique entries of the values, call set() on the collection of dictionary values as provided below. The set() returns a collection
# items, a set, with unique values, not in ordered fashion.
a =  set(fav_langs.values())

print(a)    #O/P: {'hindi', 'deutsch', 'english', 'german'}

for lang in set(fav_langs.values()):
    print(lang.title())
```

```python
#creating a set
a  = set()
a.add(1)
a.add(2)
a.add(2)    #No error returned on execution, but set doesnot add the duplicate 2
a.add(3)
a.add(4)
a.add(4)    #No error returned on execution, but set doesnot add the duplicate 4
print(a)
""" >>> a = {1,2,2,3,4,5,5}
    >>> a
    {1, 2, 3, 4, 5}
    >>> 
"""
# Unlike lists and dictionaries, sets do not retain items in any specific order.
a.pop() #Follows FIFO
print(a)
```

```python
#List in a dictionary

favorite_languages = {
    'jen': ['python', 'rust'],
    'sarah': ['c'],
    'edward': ['rust', 'go'],
    'phil': ['python', 'haskell'],
    }

for name, languages in favorite_languages.items():
    print(f"\n{name.title()}'s favorite languages are:")
    for language in languages:
        print(f"\t{language.title()}")
```
