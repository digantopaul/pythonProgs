```python
#Checking Whether a Value Is in a List using "in" Keyword

usernames = ["Aldo", "roGer", "KrishNa", "DUGGAL", "Abhik", "HULLA", "Auni"]
new_user = "KanDo"


if new_user in usernames:
    print("Username exists. Please choose a different username.")
else:
    usernames.append(new_user)

usernames
```

```python
usernames = ["Aldo", "roGer", "KrishNa", "DUGGAL", "Abhik", "HULLA", "Auni"]
new_user = "diganto"

status = 0


for user in usernames:
    if new_user.lower() == user.lower():
        status = status + 1

if status == 0:
    usernames.append(new_user)
    usernames
else:
    print("Username already exists. Please choose a different username.")

usernames.remove("diganto")
```

```python
usernames = ['Aldo', 'roGer', 'KrishNa', 'DUGGAL', 'Abhik', 'HULLA']
new_user = "Kando"

if new_user.lower() in (user.lower() for user in usernames):
    print("User exists") # (user.lower() for user in usernames) is a Generator expression. Research Generator functions, objects and expression in 
# Python. 
else:
    usernames.insert(0, new_user)
```

```python
#Checking Whether a Value Is Not in a List using "not in" keyword.

banned_users = ["Akhil", "Vinod", "Rajesh"]
user = "Sushil"

if user.lower() not in (username.lower() for username in banned_users): #again used Generator function.
    print(f"{user.title()} is not authorised for entry!")
else:
    print(f"{user.title()} is authorised for entry!")

```
