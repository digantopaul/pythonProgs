```python
#insert() method inserts element "before" the index provided in the argument, ie, the element is inserted at the mentioned index
# shifting all the elements to the right of the mentioned index one step to the right.
bikes = ['honda', 'yamaha', 'suzuki']
bikes.insert(1, "harley")

#In the above code, 'harley' was to be inserted at index 1, which shifted all the elements to the right of index 1 one more step.
bikes
```

```python
# Similarly inserting an element with a -ve index inserts the element at the negative index in the list, shifting all the elements to
#the right of the -ve index to the right by one step. Example: Here -2 index will be used by insert() method to insert "raptor" at the index -2 in the list shifting the elements to the right of 
# -2 index one step to the right.
bikes = ['honda', 'harley', 'yamaha', 'suzuki']

bikes.insert(-2,'raptor')

bikes
```

```python
#deleting element using pop() method. pop() method retains the element deleted from the list.
#By default pop() deletes the last element of the list, considering the list as a stack and performing the deletion as a LIFO.
bikes = ['honda', 'harley', 'raptor', 'ninja', 'suzuki', 'suzuki']
pop_bike = bikes.pop()
bikes
```

```python
#the last element deleted can be accessed using the variable pop_bike
pop_bike
```

```python
#adding the deleted item back to the list bikes.
bikes.append(pop_bike)
bikes
```

```python
#Popping Items from Any Position in a List. pop() method can be used to delete item from list using index number.
bikes = ['honda', 'harley', 'raptor', 'ninja', 'suzuki', 'suzuki', 'suzuki']
bikes.pop(5) 
```

```python
#Organizing a List
#sorting a list alphabetically or in ascending order
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()     #sort() method permanently alters the sequence of the items in the list.
cars
```

```python
#sorting in reverse order with sort()
cars
cars.sort(reverse=True)
cars
```

```python
#sorting a list temporarily
#to maintain the original order of the list but present it in sorted order, python provides sorted statement to sort a list. 
# Default is ascending order

cars = ['bmw', 'audi', 'toyota', 'subaru']      #creating a list
sorted(cars)        #sorting the list in ascending order
```

```python
cars                #verifying the status of the list as still maintained as original.
```

```python
#reversing the order of the list, ie in descending order with sorted() method.

sorted(cars, reverse=True)
```

```python
#verifying the status of the list
cars
```

```python
#Printing a List in Reverse Order with reverse() method. This method reverses the order of the elements in the list. it does not sort \
#the list in ascending or descending order.ÃŸ
cars.reverse()      # reverse() method reverses the order of the list permanently. Also it does not throw any output to stdout.
```

```python
cars
```

```python
#restore the previous order.
cars.reverse()
cars
```

```python
#Length of a list can be found using the len statement of python
len(cars)
```
