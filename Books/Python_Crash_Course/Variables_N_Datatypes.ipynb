```python
lang = " this is python "
print(f"hi{lang}hello")
#using strip()
print(f"hi{lang.strip()}hello", end="\n")

#using lstrip()
print(f"hi{lang.lstrip()}hello!", end="\n")

#using rstrip()
print(f"hi{lang.rstrip()}hello!", end="\n")
```
